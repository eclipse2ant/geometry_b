# 接続　-general case-

- さて一般の場合の接続について考えてみましょう。第9回の授業で、$U(1)$ の時の話をしました。この場合、接続は光子の場をあたえる。古典的には電磁気におけるベクトルポテンシャル等にあたり、曲率形式が電磁場を与えましたが、この量子化が量子電磁力学（QED）です。もちろん、量子化の手続きは、第二量子化、いわゆる場の量子論まで必要です。量子力学では、代表的なのはシュレディンガーの波動方程式ですが、これは非相対論的です。つまり特殊相対論が要求するローレンツ不変性を満たしません。相対論の要請を満たすのはクラインゴルドン方程式です、だけどこれは、ボーズ粒子にしか適用できない。電子とか陽子とかの質量をもたっような、あるいはスピンを考慮する場合、フェルミ粒子の満たす方程式、ディラック方程式が必要です。しかし、これはまだ粒子の生成消滅が記述できません。そのためには、量子力学の状態関数自体をオペレーターとして非可換化する必要があって、そこで第二量子化が行われ、場の量子論が必要になります。その場合にはファインマンの経路積分とか相関関数とか、でてきます。ただ、数学で扱う際に直接そういう話はあまり出て来ません。ある部分だけにスポットをあてて見てるようで、またファインマンの経路積分は数学的には正当化できていないこともあるかもしれません。

- $U(1)$ の時は、Maxwell方程式は量子化されて、ヤンミルズ方程式になります。また、ゲージ対称性を保つために微分作用素は共偏微分に変わりました、具体的には、接続の項を付け加えて、ゲージ変換に伴う微分作用素の変化を接続の項で吸収させ形です。今回はここから一般の場合に話を進めていきます。

- まず$G=GL(r,K)$のときを考えてみましょう。この時はファイバーが$\mathbb K^r$ベクトル束を考えて、そこでの共偏微分を考えます。

- **Conn-1.jpeg** の(2.1) で共偏微分を定義します。補題2.1.2 は　$X$の回部分集合$U$ 上の共偏微分は、$X$上の共偏微分を$U$に制限したものに尽きることを示しています。層はやってないので何ともいませんがこういう性質をflabby(脆弱性）といいます。

- (2.2) のように $End(K^r)$-値１形式で決まります。この$End(K^r)$は、$G$のリー環 $\mathbb g$ です。つまり $\mathbb g$-valued 1-form つまりこれが接続です。共偏微分にゲージ変換での不変性を課す　ないしは、局所的な座標のずれは、変換関数の分のずれを補正するこのとき、自動的に接続の満たすべき関係式が出て来ます。共変偏微分を２回施したら、線形作用層に　従って　$\mathbb g$-valued 2-form がつくれます。これが曲率形式です。また 1st Chern class も定義されます。

- さてここから一般のGに行くには　同伴束と主束の関係について、触れなくてはいけません。

- 同伴束の定義は**IMG_0030.jpeg** に書きました。　ファイバーに $G$ が作用してるとき、その作用を使って、主束からファイバーをFにそして変換関数を　$G$ の $G$ への左からの積としての作用から、$F$ への $G$ の作用に置き換えたファーバー束ができます。特に、$F$ がベクトル空間で作用が線形の時がベクトル束です。この対応を考えれば、ベクトル束で構造群 $G$ の場合の接続や共変微分の定義もわかると思います。 接続への $G$ の作用が可換の時と非可換の時でどうかわったか注意してください。物理の場合、なぜコンパクト群だけ考えるかは、リー群の構造論、つまりカルタン分解とかでユークリッドパートは可縮であるとか、もっと別の知識が必要になります。
