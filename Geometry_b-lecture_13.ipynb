 # リーマン計量

- 接ベクトル空間に内積が与えられた時、リーマン多様体といいます。$T^*(M)\otimes T^*(M)$ のsectionで、各点で正定値な対称双線型形式になってるものを与えるわけですね。もともと曲面論はガウスから始めって、測量するのに曲面論を微積を使って初めて行ったのが微分幾何のはじまり、その際、第一基本量や第二基本量を定義しますが、第一基本量がリーマン計量ですが、2点の距離はそれを結ぶ曲線の長さで測って、最短の長さを実現する曲線を測地線といいます。測地線は、その速度ベクトルが並行ベクトル場になっているもの、実は計量から接続が一意的に決まります。これをレビ・チビタ接続といいますが、接ベクトルに接続を考えれば、共偏微分でその断面、つまり接ベクトルがその接続で平行移動になってる曲線が測地線になります。

- ガウスの時代、まだカントの純粋性理性批判で、ユークリッド幾何が先天的に真であるというドグマがはびこっており、非ユークリッド幾何を考えるのはタブーな時代ですが、ガウスは定曲率の正と負の場合がボヤイとロバンチェスキーの幾何を与えることは、知ってたはずで、彼ははるか先を行ってたたわけです。そのガウス曲率は、実は第一基本量だけによるという驚くべき結果がガウスもびっくりの有名な定理です。これは、レビ・チビタ接続から決まる曲率形式を考えれば、曲率は外の世界に関係なく、曲面自体が持つ性質なわけです。これは宇宙を知る際にも、宇宙が曲がっているというのに宇宙の外に出て見なくてもよいことにもなります。この性質を2次元だけでなく高次元で推し進めたのがリーマンで、それがリーマン幾何です。それがさらにカルタンによって接続の幾何へと発展していくわけですね。

- レビ・チビタ接続の話で、特筆すべきは、トーションって概念です、代数の方に同じ言葉がありますが別物だと思います。こちらは、加群なんかの自由加群でない部分ないしは、ある加群をテンソル積をとる関手の導来関手つまりどれくらい完全性が崩れるかのホモロジーとかです。今話すトーションはTangent bundle の際に定義できる概念で、レビ・チビタ接続を定義する際にも使われます。参考文献に挙げた今野氏の本にも書かれてます。

- ブラックホールと時空の方程式　15歳からの一般相対論　　はこれで一通り読めるはずです。相対論ではリーマン計量をローレンツ計量、つまり特殊相対論は、ユークリッド幾何をミンコフスキー空間よしてローレンツ内積を考えたわけだから、その分の変更だけが一般相対論では必要です。

- ここで、物理では、一般相対論及びその量子化つまり量子重力理論へ行くか、それ以外の力、電弱相互理論だったワインバーグ・サラムに、強い力の量子色力学をいれて標準模型理論があります。これはゲージ理論ですね。一般相対論も重力子、つまりグラビトンが接続の場で記述される、ヤン・ミルズ場なのでしょうけど、繰り込みができません。ゲージ粒子は接続の場ですから接続の場全体のモジュライ空間、あるいはヤン・ミルズ場の解、特に数学ではインスタントん解のモジュライ空間（ゲージ群で割ったもの）とかが研究対象になります。この辺とドナルドソン不変量が絡んできます。

### クラインのエルランゲン・プログラム

- もうちょっと初等的な話をしておくとクラインのエルランゲン・プログラムというのがあります。ユークリッド幾何を考えましょう。形とは何か？　哲学的になっちゃいますが、二つの図形を重ね合わせて同じになるって定義すると。一つの図形をもう一つの図形に移動させないといけない。移動とは？　回転、平行、もう一つは裏表ひっくり返す。これだけです。平行移動以外は、線型変換、直交行列での変換です。行列式が$-1$ の時は回転と向きを変える変換と二つ行ってます。平行移動は${\mathbb R}^n$で表せて、二つの群の半直積であらわされます。これを合同変換群っていいます。つまり、形は、この合同変換群の作用で不変です。でも形ってわけわかんないですから、逆に開き直って、この群の変換で不変なものが形だと言い切ってしまってもいいはずです。そうすると群を変えると、いろんな不変量がでてきて新しい幾何学が生まれます。この立場がクラインのエルランゲン・プログラムです。

- 群を意識しなくても、ある種の変化に不変な性質を注目することは幾何学では重要です。また抽象化することでいろんな場面で幾何的に見ることによって、その不変量を使って現象を説明することができます。幾何学の面白いところはそこでしょう。数論幾何は数論を幾何的、どくに代数幾何学を使って数論を見る、それが現代数学をひぱっています。グロタンディックにはじまるこの動きは非常に大きいです。もう一つはホモトピー的に見る。これを代数幾何に適応する話が$A^1$ ホモトピー これによってミルナー予想が解決されました。Vladimir Voevodsky によるMotivic cohomology 理論とかそこには、ホモトピー論をカテゴリー化したモデル圏の話が関わってきます。　超弦理論においてもホログラフィック原理など次元を超えた不変性が出て来ます。そこにはホモトピー論が関わってくるように思われます。
