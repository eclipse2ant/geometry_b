# Lie 群 と Lie 環

- さてもう少しベクトルの話をしたいけど、定義は書くと煩雑になるので、ここでは違った角度から進めていきます。接束や余接束およびそのテンソル積や交代積、あるいは引き戻し等でできるベクトル束はファイバを張り合わせるときは、線形空間であるファイバーを線形変換で張り合わせました。その際に出てきたのが $GL(n,\mathbb{R})$でした。正則なn次正方行列全体ですが、行列の積に関して群になっており、$\mathbb{R}^{n^2}$ の開集合になっており多様体にもなっているのでした。行列の積と逆行列をとる写像は微分可能写像になっており、こういう群をリー群といいます。リーは数学者の名前です。これについては 参考資料のトピックにある **IMG_0015** を見てください。 

- $GL(n,\mathbb{R})$ は一般線形群と呼ばれます。他の例は **IMF_0016** にあります。複素の場合、複素多様体で積と逆元をとる写像が正則な群を複素リー群といいます。$GL(n,\mathbb{C})$ を 一般複素線形群といい複素リー群です。$O(n), U(n)$ は直交行列からなる群　ユニタリ行列からなる群で、それぞれ直交群、ユニタリ群で　一般線形群、一般複素線形群の部分群になっていてさらにはコンパクト群です。 

- $SL(n,\mathbb{R}),\  SL(n,\mathbb{C})$ の定義は IMG_0016 を見てください。特殊線形群、特殊複素線形群といいます。これらの多様体としての局所座標の入れ方は院関数定理を使って、埋めこまれる部分多様体としての議論を使います。ただしここではそこには触れません。あと特殊直交群、特殊ユニタリ群 $SO(n),\  SU(n)$ といったコンパクト群もあります。確か位相群で局所コンパクトならばリー群になったとおもうのですが、ここで扱う例は、線形リー群とよばれる、行列で表される例ばかりしか扱いません。


- Def. 1.10.2 にある　$L_g,\ R_g$ は微分可能な全単射な写像です。リー群を$G$ とすると、単位元$e$での接空間 $T_e(G)$ を $\mathbb{g}$ で表し、$G$のリー環またはリー代数といいます。　$\bar{X}$ を$L_g$の微分写像 $(L_g)_*$ を使って定義して　ベクトル場を定義します。この場に関して　交代積で $[\bar{X},\bar{Y}]_e$ で  $[X,Y]$ を定義します。IMG_0037 に書きましたが、一般線形群の時、そのりー環 gl(n,$\mathbb{R})$ は、$Mat(n,R)$ になります。この辺を言うには行列の指数関数写像　$exp\ X$ を使います　$exp\ tX$ の$t=0$ の接ベクトル(速度ベクトル）は $X$ です。　また、基本ベクトル場というのはあとで使います。

# 主ファイバー束とゲージ変換

- IMG_0016-17 に**主ファイバー束（Principal Fiber Bundle)** の定義を書きました。今度はファイバー自体が$G$ で　$U\times G$ たちを$G$ で張り合わせるという風になります。どういう風に張り合わせるか、今までの接束などの例と比較してみてください。主ファイバー束の例として IMG_0018 にフレーム束、多様体の各点にその接空間の基底の全体をファイバーとして張り付けたもの、基底は正則行列全体で表せて、基底の変換則が張り合わせ方になっています。これと、接ベクトルとの関係はこれから出て来ます。いわゆる同伴束です。
